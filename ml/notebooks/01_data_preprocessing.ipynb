# Insights Lab: Data Preprocessing

This notebook cleans the campus placement dataset, handles missing values, and exports a cleaned CSV for model training.

```python
from pathlib import Path

import pandas as pd

DATA_PATH = Path('..') / '..' / 'frontend' / 'public' / 'campus_predictor_sample.csv'
ARTIFACTS_DIR = Path('..') / 'artifacts'
ARTIFACTS_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

df.info()
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(exclude=['number']).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

df['status'] = df['status'].astype(str).str.strip()

df['placed'] = df['status'].str.lower().str.contains('select').astype(int)
df['salary'] = df['package']

df[['status', 'placed']].head()
```

```python
cleaned_path = ARTIFACTS_DIR / 'cleaned_dataset.csv'
df.to_csv(cleaned_path, index=False)

cleaned_path
```
